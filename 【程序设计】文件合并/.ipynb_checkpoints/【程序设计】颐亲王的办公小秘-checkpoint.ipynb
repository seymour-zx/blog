```python
#!/usr/bin/python
# -*- coding: utf-8 -*-



#创建程序窗口 —— createProgramWindow
def createProgramWindow(programTitle=None,programIco=None,programWidth=None,programHeight=None):
#     导入模块
    import tkinter
#     创建
    programWindow = tkinter.Tk()
#     【标题】
    if programTitle == None:
        programWindow.title('颐亲王的办公小秘')  
    else:
        programWindow.title(programTitle)
#     【图标】
    if programIco == None:
        programWindow.iconbitmap(r'图标\图标.ico') 
    else:
        programWindow.iconbitmap(programIco) 
#     默认 【大小】 和 【位置】（'宽度 x 高度 + 偏右量 + 偏下量'）
    if programWidth == None:
        programWidth = 480
    if programHeight == None:
        programHeight = 270     
    programWindow.geometry(str(programWidth)+'x'+str(programHeight)+'+250+150')         
#     ( 行 — 列 )权重
    programWindow.rowconfigure(0,weight=1)
    programWindow.columnconfigure(0,weight=1)    
#     更新程序窗口
    programWindow.update()
#     注册程序窗口
    return programWindow


# 创建程序菜单    
def createProgramMenu(master,listMenu,menuCommand):
    import tkinter
#     创建下拉菜单
    subMenu = tkinter.Menu(master,tearoff = 0)
    for cellMenu in listMenu :
        if isinstance(cellMenu,str):
            subMenu.add_command(label = cellMenu,command = menuCommand.get(cellMenu),font=('微软雅黑',9))
        elif isinstance(cellMenu,list):
            masterCellMenu = cellMenu[0]
            del(cellMenu[0])
            subMenu.add_cascade(label = masterCellMenu ,font='微软雅黑',menu = createProgramMenu(subMenu,cellMenu,menuCommand))            
    return subMenu   


# 创建板块框体
def createFrame(master,text=None,font=None,labelanchor=None,fg=None,bg=None,row=None,column=None):
    import tkinter
    if font==None:
        font = ('微软雅黑',14)
    if labelanchor==None:
        labelanchor = 'n'
    if fg ==None:
        fg='black'
    if bg==None:
        bg='white'
    defaultFrame = tkinter.LabelFrame(master,font=font,text=text,bg=bg,fg=fg,labelanchor=labelanchor)
    if row == None:
        row = 0
    if column == None:
        column = 0
    defaultFrame.grid(row = row,column = column, padx=1, pady=1,ipadx=1, ipady=1,stick = tkinter.NSEW)
    defaultFrame.rowconfigure(0,weight=1)
    defaultFrame.columnconfigure(0,weight=1) 
    return defaultFrame


#创建按钮
def createButton(master,text=None,command=None,row=None,column=None,rowspan=None,columnspan=None ):
    import tkinter
    defaultButton = tkinter.Button(master,text=text,command=command)
    if row==None:
        row=0
    if column==None:
        column=0
    if rowspan==None:
        rowspan=1
    if columnspan==None:
        columnspan=1
    defaultButton.grid(row=row,column=column,rowspan=rowspan,columnspan=columnspan,stick = tkinter.NSEW)
    return defaultButton


#创建输入框
def createEntry(master,textvariable=None,command=None,row=None,column=None,rowspan=None,columnspan=None ):
    import tkinter
    defaultEntry = tkinter.Entry(master,textvariable=textvariable,command=command)
    if row==None:
        row=0
    if column==None:
        column=0
    if rowspan==None:
        rowspan=1
    if columnspan==None:
        columnspan=1
    defaultEntry.grid(row=row,column=column,rowspan=rowspan,columnspan=columnspan,stick = tkinter.NSEW)
    return defaultEntry
    
    
#创建标签
def createLabel(master,text=None,command=None,row=None,column=None,rowspan=None,columnspan=None ):
    import tkinter
    defaultLabel = tkinter.Label(master,text=text,command=command)
    if row==None:
        row=0
    if column==None:
        column=0
    if rowspan==None:
        rowspan=1
    if columnspan==None:
        columnspan=1
    defaultLabel.grid(row=row,column=column,rowspan=rowspan,columnspan=columnspan,stick = tkinter.NSEW)
    return defaultLabel




#【程序主要区域】
if __name__ == "__main__":
        
    
#     导入全局模块
    import tkinter

# 【程序主视窗】    
#     创建程序视窗
    programWindow = createProgramWindow( programTitle=None, programIco=None, programWidth=None, programHeight=None )
    
    
#     创建程序菜单栏
#         程序菜单名列表
    listMenu = [['File','Open',"Exit"],'Edit','View','Insert','Cell','Kernel','Help']
#         程序菜单命令字典
    menuCommand = {"Exit":lambda:programWindow.destroy()}
#         根据列表和字典给程序添加主菜单栏
    programWindow.config( menu = createProgramMenu( programWindow, listMenu, menuCommand ) )
    
    
#     创建程序主板块
    programFrame = createFrame(programWindow,bg='#008080')
    programFrame.columnconfigure(1,weight=6)       
#     左板块【按钮区域】
    leftFrame = createFrame(programFrame)      
#     右板块【显示区域】
    rightFrame = createFrame(programFrame,column =1)
    
     
    
#左板块【按钮区域】leftFrame
    if True:
        
#打开计算器        
        buttonCalculator = createButton(leftFrame,text='计算器',command=lambda:openCalculator())
#退出程序
        buttonProgramExit = createButton(leftFrame,row =1,text='退出程序',command=lambda:programWindow.destroy())        



#右板块【显示区域】rightFrame
    if True:
        
#打开计算器        
        def openCalculator():
        
#     创建计算器板块
            calculatorFrame = createFrame(rightFrame)
#     调整计算器板块内的( 行 — 列 )权重
            for x in range(8):
                calculatorFrame.rowconfigure(x,weight=1)
            for x in range(6):
                calculatorFrame.columnconfigure(x,weight=1)


#     创建计算器显示屏幕      
            calculatorEntry = createEntry(calculatorFrame,row=1,column=1,columnspan=4)


#     设置命令按钮

#         关闭计算器                 
            buttonCloseCalcultor = createButton(calculatorFrame,row=0,column=5,text ='CloseCalculator',command=lambda:calculatorFrame.destroy())

#         清空计算器显示屏幕                 
            buttonClearCalcultor = createButton(calculatorFrame,row=2,column=1,text ='C',command=lambda:calculatorEntry.delete(0,'end'))

#         退格               
            buttonBackSpaceCalcultor = createButton(calculatorFrame,row=2,column=4,text ='退格',command=lambda:calculatorEntry.delete(len(calculatorEntry.get())-1,'end'))
    

#         小数点
#             默认未输入过小数点
            numberPoint=False
#             创建按钮【小数点】
            buttonNumberPoint = createButton(calculatorFrame,text=".",row =6,column=3,command=lambda :increaseNumberPoint(calculatorEntry))    
#             【小数点】按钮命令
            def increaseNumberPoint(calculatorEntry):
                global numberPoint
#                 判断是否未输入过【小数点】
#                     未输入则执行
                if not numberPoint:
#                     在窗口末尾添加【小数点】
                    calculatorEntry.insert('end','.')
#                     标记为添加过【小数点】
                    numberPoint=True

#         数字 1 - 9
            for x in range(1,10):
                buttonNumber = createButton(calculatorFrame,text=x,row=5-((x-1)//3),column=1+(x-1)%3,command=lambda i=x:calculatorEntry.insert('end',i))

#         数字 0
            number0=createButton(calculatorFrame,text="0",row =6,column=1,columnspan=2,command=lambda :calculatorEntry.insert('end',0))

#         运算符号 "+" "-" "x" "/"
            calculatingSign=False
            buttonPlus = createButton(calculatorFrame,text="+",row =3,column=4,command=lambda :calculatingSigns("+",calculatorEntry))
            buttonMinus = createButton(calculatorFrame,text="-",row =4,column=4,command=lambda :calculatingSigns("-",calculatorEntry))
            buttonMultiply = createButton(calculatorFrame,text="x",row =2,column=2,command=lambda :calculatingSigns("x",calculatorEntry))
            buttonDivided = createButton(calculatorFrame,text="/",row =2,column=3,command=lambda :calculatingSigns("/",calculatorEntry))
#             【运算符号】按钮命令
            def calculatingSigns(sign,calculatorEntry):
                global calculatingSign,calculatingString1,numberPoint
#                 判断是否未输入过【运算符号】
#                     未输入则执行
                if calculatingSign==False:
#                     保存【运算符号】
                    calculatingSign=sign
#                     保存第一个运算数字
                    calculatingString1=calculatorEntry.get()
#                     清空输入框
                    calculatorEntry.delete(0,"end")
#                     重置【小数点】
                    numberPoint=False

#         运算符号 "="                
            buttonEqual=createButton(calculatorFrame,text='=',row =5,rowspan=2,column=4,command=lambda :calculatingEqual(calculatorEntry))               
#             【"="】按钮命令         
            
            def calculatingEqual(calculatorEntry):
                global calculatingSign,calculatingString1,numberPoint
#                 判断是否未输入过【运算符号】
#                     输入过则执行                
                if calculatingSign:  
                    String1=calculatingString1.split('.')   
                    if len(String1)==0:
                        str1=0
                    elif len(String1)==1:
                        if String1[0]=="":
                            str1=0
                        else:
                            str1=int(String1[0])
                    else:
                        if String1[0]=="":
                            if String1[1]=="":
                                str1=1
                            else:
                                str1 = int(String1[1])*0.1**len(String1[1])
                        else:
                            if String1[1]=="":
                                str1=int(String1[0])
                            else:
                                str1 = int(String1[0])+int(String1[1])*0.1**len(String1[1])
                    calculatingString2=calculatorEntry.get()
                    String2=calculatingString2.split('.')    
                    if len(String2)==0:
                        str2=0
                    elif len(String2)==1:
                        if String2[0]=="":
                            str2=0
                        else:
                            str2=int(String2[0])
                    else:
                        if String2[0]=="":
                            if String2[1]=="":
                                str2=1
                            else:
                                str2 = int(String2[1])*0.1**len(String2[1])
                        else:
                            if String2[1]=="":
                                str2=int(String2[0])
                            else:
                                str2 = int(String2[0])+int(String2[1])*0.1**len(String2[1])   
                            
                    if calculatingSign=="+":
                        str3 =  str1 + str2
                    if calculatingSign=="-":
                        str3 =  str1 - str2    
                    if calculatingSign=="x":
                        str3 =  str1 * str2            
                    if calculatingSign=="/":
                        str3 =  str1 / str2  
                                   
                    calculatorEntry.delete(0,"end")
                    calculatingString1=str(str3)
                    calculatorEntry.insert(0,calculatingString1)
                    calculatingSign=False
                    numberPoint=False

#     frame11.config(text= 'ddddd')
    
#     程序窗口隐藏 1 秒后显示
#     rootWindow.withdraw()
#     import time
#     time.sleep(1)
#     rootWindow.deiconify()    
    
    programWindow.update()
    programWindow.mainloop()
```

```python
import tkinter
#创建程序视窗
programWindow = tkinter.Tk()
#设置程序名称
programWindow.title('颐亲王的办公小秘')  
#设置程序图标
programWindow.iconbitmap(r'图标\图标.ico') 
#设置程序起始【大小】 和 【位置】
  #单位：像素（'宽度 x 高度 + 偏右量 + 偏下量'）    
programWindow.geometry('480x270+250+150')       
#设置【行|列】权重（权重weight默认值为'0'）
#将第1行的权重值更改为'1'
programWindow.rowconfigure(0,weight=1)
#将第1列的权重值更改为'1'
programWindow.columnconfigure(0,weight=1) 
#背景设置为【黑色】
programWindow.configure(bg='black')
#更新程序窗口
programWindow.update()
#注册程序窗口
programWindow.mainloop()
```
