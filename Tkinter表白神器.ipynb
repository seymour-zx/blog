```python
import webbrowser #导入"网页&浏览器“模块
import tkinter.messagebox #导入"消息提示窗口“模块
from tkinter import * #导入"tkinter“模块
from PIL import ImageTk,Image #导入"图片处理“模块
from itertools import count #导入"计帧数工具“模块
import os, sys

# #对象使用内存大小
# #下面的代码片段展示了，如何获取一个对象所占用的内存大小：
# import sys
# num = '0.9'
# print(sys.getsizeof(num))

def resource_path(relative):
    if hasattr(sys, "_MEIPASS"):
        absolute_path = os.path.join(sys._MEIPASS, relative)
    else:
        absolute_path = os.path.join(relative)
    return absolute_path


class ImageLabel(Label):
    """a label that displays images, and plays them if they are gifs"""
    def load(self, im):
        if isinstance(im, str):
            im = Image.open(im)
        self.loc = 0
        self.frames = []

        try:
            for i in count(1):
                self.frames.append(ImageTk.PhotoImage(im.copy()))
                im.seek(i)
        except EOFError:
            pass

        try:
            self.delay = im.info['duration']
        except:
            self.delay = 100

        if len(self.frames) == 1:
            self.config(image=self.frames[0])
        else:
            self.next_frame()
            
    def unload(self):
        self.config(image="")
        self.frames = None
        
    def next_frame(self):
        if self.frames:
            self.loc += 1
            self.loc %= len(self.frames)
            self.config(image=self.frames[self.loc])
            self.after(self.delay, self.next_frame)
            
            
def Uniform_Resource_Locator():
    url=webbrowser.open_new("http://www.baidu.com")
    #在浏览器的新页面，打开制定网址

    
def goodbye1():
    messagebox.askokcancel("Hello !","Good Bye !")
    messagebox.askyesno("Hello !","Good Bye !")
    messagebox.askretrycancel("Hello !","Good Bye !")
    messagebox.showinfo("Hello !","Good Bye !")
    messagebox.showerror("Hello !","Good Bye !")    
    messagebox.showwarning("Hello !","Good Bye !")
    #关闭父窗口,会死机，别用！！！
    #root.withdraw()

    
def callback1():
    entry = Entry(root,bd=5,textvariable = var)
    var.set('请告诉我你的回答')
    entry.grid(row=3,column=0)#在窗口展示entry输入框对象
    
    

root = Tk() #调用Tk（）函数创建根窗口
root.title('你喜欢我吗？') #设置根窗口标题
root.geometry('400x400+250+220') #设置根窗口大小

l1 = Label(root,text = 'hey,girly girl',font=('微软雅黑',14),fg="white",bg="#008080")
l1.grid(row=0,column=0)#在窗口展示Label标签对象

var = StringVar()
l2 =Label(root,textvariable = var ,font=('微软雅黑',14))
var.set('当然！\n我非常喜欢你！')
#\n 强制换行
l2.grid(row=1,column=0)#在窗口展示Label标签对象

stud1=Button(root, text="一起吃饭吗？",fg='red',bd = 2,width =28,command=callback1 )
stud1.grid(row=2,column=0)#在窗口展示Button按钮对象

stud2=Button(root, text="上网搜索答案", font=('Courier',15,'bold'), command=Uniform_Resource_Locator)
stud2.grid(row=4,column=0)#在窗口展示Button按钮对象

stud3=Button(root, text="Say goodbye !", command= goodbye1 )
stud3.grid(row=5,column=0)#在窗口展示Button按钮对象

l3 = ImageLabel(root)
l3.grid(row= 6, column = 0)
l3.load(resource_path('./3.gif'))

root.mainloop()#保持窗口持续展示
```
