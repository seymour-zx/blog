```python
#!/usr/bin/python
# -*- coding: utf-8 -*-
import os, sys
sys.modules
os.environ

if __name__ == "__main__":
    import pandas
    from datetime import datetime
    dfs = pandas.read_html("https://rate.bot.com.tw/xrt?Lang=zh-TW")
    currency = dfs[0]
```

```python
#行全部，列前5
currency = currency.ix[:,0:5]
currency.columns = [u"币别",u"现金汇率 - 本行买入",u"现金汇率 - 本行卖出",u"即期汇率 - 本行买入",u"即期汇率 - 本行卖出"]
currency[u"币别"] = currency[u"币别"].str.extract('\((\w+)\)')
currency[u"Date"] = datetime.now().strftime("%Y-%m-%d")
currency[u"Date"] = pandas.to_datetime(currency[u"Date"])
currency
```
