```python
#!/usr/bin/python
# -*- coding: utf-8 -*-
import os, sys
sys.modules
os.environ

if __name__ == "__main__":
    import requests
     #打开ip地址列表网站   
    res = requests.get("https://free-proxy-list.net/")
```

```python
import re
#筛选Ip地址
m = re.findall("\d+\.\d+\.\d+\.\d+:\d+",res.text)
#储存有效地址
validips = []

for ip in m:
    try:
        #使用代理IP地址访问测试网站
        res =requests.get("https://api.ipify.org?format=json",proxies = {"http":ip,"https":ip},timeout = 1)
        validips.append({"ip":ip})
        print(res.json())        
    except:
        print("Failed",ip)
```
